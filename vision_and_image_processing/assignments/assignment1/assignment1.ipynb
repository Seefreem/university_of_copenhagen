1. An inner and an outer product. Compute the inner product

```python
a = np.array([17,6,-5.5,3]) 
b = np.array([-4.9,1.17,0,2])
```

```python
# inner product of a and b is:
print(a.dot(b))
print(np.dot(a, b))
print(a @ b)
```

```python
# outer product of a and b is :
a = a.reshape((a.shape[0], 1))
b = b.reshape((b.shape[0], 1))
# print(a.shape, b.shape)
print(a@b.T)
```

2. Let t = [−3, 2, 11]T . Compute the cross-product

```python
# my implementation of cross product:
def CrossProduct(a, b):
    return np.array([a[1]*b[2] - a[2]*b[1], -(a[0]*b[2] - a[2]*b[0]), a[0]*b[1] - a[1]*b[0]])
```

```python
t = np.array([-3, 2, 11])
print('the cross-product t × t is:', np.cross(t, t))
print('the cross-product t × t is:', CrossProduct(t, t))
```

```python
a = np.array([-2, 6, 1])
b = np.cross(t, a)
c = np.cross(a, t)
print('b: ', b)
print('c: ', c)
print('tTb: ', t.T@b)
print('aTb: ', a@b)
```

```python
X = np.arange(-7, 7., 0.1)
Y = [f(x) for x in X]
Y_1 = [f_prime(x) for x in X]
Y_2 = [f_double_pirme(x) for x in X]
plt.plot(X, Y, label='f(x)')
plt.plot(X, Y_1, label='f\'(x)')
plt.plot(X, Y_2, label='f\'\'(x)')
plt.legend()
```

5. Plot f (x, t)

```python
X = np.arange(-15, 15, 0.1)

def gd_function(x, t):
    return 1 / np.sqrt((2 * np.pi * t)) * pow(np.e, -1 * x * x / 2 / t)

Y = [gd_function(x, 1) for x in X]
Y_2 = [gd_function(x, 2) for x in X]
Y_4 = [gd_function(x, 4) for x in X]

plt.plot(X, Y, label='t = 1')
plt.plot(X, Y_2, label='t = 2')
plt.plot(X, Y_4, label='t = 4')
plt.title('Gaussian distribution')
plt.legend()
```

6. Plot ∂f/∂x (x, t)

<img src="./images/3.6.png">

```python
def gd_function_dx(x, t):
    return -1 * x / t * gd_function(x, t) 

Y_dx = [gd_function_dx(x, 1) for x in X]
Y_2_dx = [gd_function_dx(x, 2) for x in X]
Y_4_dx = [gd_function_dx(x, 4) for x in X]

plt.plot(X, Y_dx, label='t = 1')
plt.plot(X, Y_2_dx, label='t = 2')
plt.plot(X, Y_4_dx, label='t = 4')
plt.title('x-derivative of Gaussian distribution')
plt.legend()
```

7. Plot the t-derivative of f (x, t)

<img src="./images/3.7.png">

```python
def gd_function_dt(x, t):
    return (x * x / 2 / t / t - 1 / 2 / t) * gd_function(x, t) 

Y_dt = [gd_function_dt(x, 1) for x in X]
Y_2_dt = [gd_function_dt(x, 2) for x in X]
Y_4_dt = [gd_function_dt(x, 4) for x in X]

plt.plot(X, Y_dt, label='t = 1')
plt.plot(X, Y_2_dt, label='t = 2')
plt.plot(X, Y_4_dt, label='t = 4')
plt.title('t-derivative of Gaussian distribution')
plt.legend()
```
